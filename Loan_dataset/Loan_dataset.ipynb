#  Load the Dataset

```python
# import libarary
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

loan=pd.read_csv("Loan_dataset.csv")

# first few rows of the dataset
loan.head()
```

#  Summary Statistics

```python
loan.describe()
```

```python
loan.info() 
```

# Analysis and Visualization

### 1. Average Current Loan Amount for Each Loan Status

```python
average_loan_amount = loan.groupby('Loan Status')['Current Loan Amount'].mean()
print("Average current loan amount for each loan status:\n", average_loan_amount)
```

### 2. Credit Score vs Annual Income

```python
plt.scatter(data=loan,x='Annual Income',y='Credit Score')
plt.xlabel('Annual Income')
plt.ylabel('Credit Score')
plt.title('credit score vary with the annual income')
plt.show()
```

### 3. Correlation Between Number of Open Accounts and Current Credit Balance

```python
correlation_open_accounts_balance = loan[['Number of Open Accounts','Current Credit Balance']].corr().iloc[0,1]
print("Correlation between number of open accounts and current credit balance:", correlation_open_accounts_balance.round(2))


sns.scatterplot(x=loan['Number of Open Accounts'], y=loan['Current Credit Balance'])
plt.xlabel('Number of Open Accounts')
plt.ylabel('Current Credit Balance')
plt.title('Correlation Between Number of Open Accounts and Current Credit Balance')
plt.show()
```

### 4. Distribution of Credit Scores Across Different Home Ownership Types

```python
credit_score_distribution = loan.groupby('Home Ownership')['Credit Score'].describe()
print("Distribution of credit scores across different home ownership types:")
pd.DataFrame(credit_score_distribution)
```

```python
sns.boxplot(x='Home Ownership', y='Credit Score', data=loan)
plt.xlabel('Home Ownership')
plt.ylabel('Credit Score')
plt.title('Distribution of Credit Scores Across Home Ownership Types')
plt.show()
```

### 5. Annual Income by Loan Purpose

```python
annual_income_by_purpose = loan.groupby('Purpose')['Annual Income'].mean()
print("Annual income for different purposes of loans:\n", annual_income_by_purpose)
```

```python
sns.boxplot(x=loan['Purpose'], y=loan['Annual Income'])
plt.xlabel('Purpose')
plt.ylabel('Annual Income')
plt.title('Annual Income by Loan Purpose')
plt.xticks(rotation=90)
plt.show()
```

### 6. Average Monthly Debt for Each Term (Short-term vs. Long-term)

```python
average_monthly_debt = loan.groupby('Term')['Monthly Debt'].mean()
print("Average monthly debt for each term:\n", average_monthly_debt)
```
